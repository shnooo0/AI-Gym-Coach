```python
import os
import cv2
import numpy as np
from tensorflow.keras.applications import MobileNetV2
from tensorflow.keras.applications.mobilenet_v2 import preprocess_input
from tensorflow.keras.models import Model
```

```python
# Load PoseNet model
base_model = MobileNetV2(weights='imagenet', include_top=False, input_shape=(224, 224, 3))
base_model.trainable = False
```

```python
# Extract features from a given image using PoseNet
def extract_pose_features(image_path, model):
    # Load and preprocess the image
    image = cv2.imread(image_path)
    image = cv2.resize(image, (224, 224))
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    preprocessed_image = preprocess_input(image)
    preprocessed_image = np.expand_dims(preprocessed_image, axis=0)

    # Extract features using PoseNet
    features = model.predict(preprocessed_image)

    return features.flatten()  # Flatten the features for simplicity
```

```python
# Function to process images in a directory and save the extracted features
def process_images_in_directory(input_dir, output_dir, model):
    if not os.path.exists(output_dir):
        os.makedirs(output_dir)

    for filename in os.listdir(input_dir):
        if filename.endswith('.jpg') or filename.endswith('.png'):
            image_path = os.path.join(input_dir, filename)
            features = extract_pose_features(image_path, model)
            output_filename = os.path.join(output_dir, filename.split('.')[0] + '.npy')
            np.save(output_filename, features)
```

```python
# Specify input and output directories
input_dataset_dir = 'OneDrive/Desktop/Dataset/deadlift'
output_dataset_dir = 'OneDrive/Desktop/PoseDataset/Deadlift'
```

```python
# Process images in the input directory and save features to the output directory
process_images_in_directory(input_dataset_dir, output_dataset_dir, base_model)
```
